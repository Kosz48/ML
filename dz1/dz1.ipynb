```python
import pandas as pd
import numpy as np
import os

# Путь к данным
data_dir = '/kaggle/input/human-activity-recognition-with-smartphones/'

# Загрузка тренировочных данных
train_data = pd.read_csv(os.path.join(data_dir, 'train.csv'))

# Загрузка тестовых данных
test_data = pd.read_csv(os.path.join(data_dir, 'test.csv'))

print("Данные успешно загружены!")
```

```python
# Просмотр первых строк данных
print("Первые 5 строк тренировочных данных:")
display(train_data.head())

print("\nИнформация о тренировочных данных:")
print(train_data.info())

print("\nСтатистические характеристики тренировочных данных:")
display(train_data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Распределение классов в тренировочных данных
plt.figure(figsize=(8,6))
sns.countplot(x='Activity', data=train_data)
plt.title('Распределение классов активности в тренировочных данных')
plt.xticks(rotation=45)
plt.show()
```

```python
# Вычисление корреляционной матрицы
corr_matrix = train_data.drop(['Activity', 'subject'], axis=1).corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Корреляционная матрица признаков')
plt.show()
```

```python
# Проверка наличия пропущенных значений
print("Количество пропущенных значений в тренировочных данных:")
print(train_data.isnull().sum())

print("\nКоличество пропущенных значений в тестовых данных:")
print(test_data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Преобразование меток в числовой формат
train_data['Activity'] = le.fit_transform(train_data['Activity'])
test_data['Activity'] = le.transform(test_data['Activity'])

# Сохранение соответствия классов и их кодов
activity_labels = dict(zip(le.classes_, le.transform(le.classes_)))
print("Коды классов активности:")
print(activity_labels)
```

```python
from sklearn.preprocessing import StandardScaler

# Выделение признаков и меток
X_train = train_data.drop(['Activity', 'subject'], axis=1)
y_train = train_data['Activity']

X_test = test_data.drop(['Activity', 'subject'], axis=1)
y_test = test_data['Activity']

# Проверка размеров
print("Размер X_train:", X_train.shape)
print("Размер y_train:", y_train.shape)
print("Размер X_test:", X_test.shape)
print("Размер y_test:", y_test.shape)

# Нормализация признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Проверка среднего и стандартного отклонения первых 5 признаков
print("\nСреднее значение первых 5 признаков после масштабирования:")
print(np.mean(X_train_scaled[:, :5], axis=0))

print("\nСтандартное отклонение первых 5 признаков после масштабирования:")
print(np.std(X_train_scaled[:, :5], axis=0))

from sklearn.model_selection import train_test_split

# Разбиение тренировочных данных на обучающую и валидационную выборки
X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train_scaled, y_train, test_size=0.2, random_state=42, stratify=y_train)

# Проверка размеров после разбиения
print("\nРазмер обучающей выборки X_train_split:", X_train_split.shape)
print("Размер валидационной выборки X_val:", X_val.shape)
print("Размер обучающей метки y_train_split:", y_train_split.shape)
print("Размер валидационной метки y_val:", y_val.shape)

# Проверка распределения классов
print("\nРаспределение классов в y_train_split:")
print(y_train_split.value_counts())

print("\nРаспределение классов в y_val:")
print(y_val.value_counts())
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Создание экземпляра модели с k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Обучение модели
knn.fit(X_train_split, y_train_split)
```

```python
from sklearn.model_selection import cross_val_score

k_range = range(1, 21)
cv_scores = []

# Поиск оптимального k
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train_split, y_train_split, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())

# Определение значения k с наибольшей точностью
optimal_k = k_range[np.argmax(cv_scores)]
print(f"Оптимальное значение k: {optimal_k}")

# График зависимости точности от k
plt.figure(figsize=(10,6))
plt.plot(k_range, cv_scores, marker='o')
plt.xlabel('Значение k для k-NN')
plt.ylabel('Кросс-валидационная точность')
plt.title('Подбор оптимального k')
plt.show()
```

```python
# Обучение модели с оптимальным k
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(X_train_split, y_train_split)

# Предсказания на обучающей, валидационной и тестовой выборках
y_train_pred = knn_optimal.predict(X_train_split)
y_val_pred = knn_optimal.predict(X_val)
y_test_pred = knn_optimal.predict(X_test_scaled)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Отчет по обучающей выборке
print("Отчет по обучающей выборке:")
print(classification_report(y_train_split, y_train_pred))

# Отчет по валидационной выборке
print("Отчет по валидационной выборке:")
print(classification_report(y_val, y_val_pred))

# Отчет по тестовой выборке
print("Отчет по тестовой выборке:")
print(classification_report(y_test, y_test_pred))

# Вычисление точности
train_accuracy = accuracy_score(y_train_split, y_train_pred)
val_accuracy = accuracy_score(y_val, y_val_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Точность на обучающей выборке: {train_accuracy:.4f}")
print(f"Точность на валидационной выборке: {val_accuracy:.4f}")
print(f"Точность на тестовой выборке: {test_accuracy:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Обучение модели случайного леса
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_split, y_train_split)

# Предсказания
y_val_pred_rf = rf.predict(X_val)
y_test_pred_rf = rf.predict(X_test_scaled)

# Отчет по валидационной выборке
print("Отчет по случайному лесу на валидационной выборке:")
print(classification_report(y_val, y_val_pred_rf))

# Точность на валидационной выборке
val_accuracy_rf = accuracy_score(y_val, y_val_pred_rf)
print(f"Точность случайного леса на валидационной выборке: {val_accuracy_rf:.4f}")
```

```python
from sklearn.svm import SVC

# Обучение модели SVM
svc = SVC(kernel='rbf', C=1, gamma='scale', random_state=42)
svc.fit(X_train_split, y_train_split)

# Предсказания
y_val_pred_svc = svc.predict(X_val)
y_test_pred_svc = svc.predict(X_test_scaled)

# Отчет по валидационной выборке
print("Отчет по SVM на валидационной выборке:")
print(classification_report(y_val, y_val_pred_svc))

# Точность на валидационной выборке
val_accuracy_svc = accuracy_score(y_val, y_val_pred_svc)
print(f"Точность SVM на валидационной выборке: {val_accuracy_svc:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ваши данные
models = ['k-NN', 'Random Forest', 'SVM']
accuracies = [val_accuracy, val_accuracy_rf, val_accuracy_svc]

# Преобразуем списки в numpy массивы
models_array = np.array(models)
accuracies_array = np.array(accuracies)

# Построение графика
plt.figure(figsize=(8,6))
sns.barplot(x=models_array, y=accuracies_array)
plt.ylabel('Точность на валидационной выборке')
plt.title('Сравнение моделей')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Применение SMOTE к обучающим данным
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_split, y_train_split)

# Обучение модели на новых данных
rf_smote = RandomForestClassifier(n_estimators=100, random_state=42)
rf_smote.fit(X_train_resampled, y_train_resampled)

# Предсказания
y_val_pred_rf_smote = rf_smote.predict(X_val)

# Отчет по валидационной выборке
print("Отчет по случайному лесу с SMOTE на валидационной выборке:")
print(classification_report(y_val, y_val_pred_rf_smote))
```

```python
import numpy as np

# Вычисление корреляционной матрицы
corr_matrix = pd.DataFrame(X_train_split).corr().abs()

# Выбор верхнего треугольника матрицы корреляции
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Нахождение признаков с высокой корреляцией (>0.95)
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

print(f"Количество признаков до удаления: {X_train_split.shape[1]}")
print(f"Количество признаков для удаления: {len(to_drop)}")

# Удаление признаков
X_train_reduced = pd.DataFrame(X_train_split).drop(columns=to_drop)
X_val_reduced = pd.DataFrame(X_val).drop(columns=to_drop)
X_test_reduced = pd.DataFrame(X_test_scaled).drop(columns=to_drop)

# Обучение модели на уменьшенном наборе признаков
rf_reduced = RandomForestClassifier(n_estimators=100, random_state=42)
rf_reduced.fit(X_train_reduced, y_train_split)

# Предсказания
y_val_pred_rf_reduced = rf_reduced.predict(X_val_reduced)

# Отчет по валидационной выборке
print("Отчет по случайному лесу на уменьшенном наборе признаков:")
print(classification_report(y_val, y_val_pred_rf_reduced))
```
